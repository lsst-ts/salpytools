```python
from salpytools import salpylib 
import logging
```

```python
logging.basicConfig(level=logging.DEBUG,
                        format='[%(asctime)s] [%(levelname)s] %(message)s',
                        datefmt='%Y-%m-%d %H:%M:%S')
```

```python
class FakeSchedulerCSC():
    
    def __init__(self):
        self.device = 'scheduler'
        self.machine = salpylib.StateMachine(self.device)
        
        self.enter_control = salpylib.DDSController('enterControl', 'scheduler', 
                                                    state_machine=self.machine)  # OFFLINE >> STANDBY
        self.start = salpylib.DDSController('start', 'scheduler', state_machine=self.machine)  # STANDBY >> DISABLE
        self.enable = salpylib.DDSController('enable', 'scheduler', state_machine=self.machine)  # DISABLE >> ENABLE
        
        self.disable = salpylib.DDSController('disable', 'scheduler', state_machine=self.machine)  # ENABLE >> DISABLE
        self.standby = salpylib.DDSController('standby', 'scheduler', state_machine=self.machine)  #  DISABLE >> STANDBY
        self.exit_control = salpylib.DDSController('exitControl', 'scheduler', 
                                                    state_machine=self.machine)  # STANDBY >> OFFLINE

        
        self.enter_control.start()
        self.start.start()
        self.enable.start()
        self.disable.start()
        self.standby.start()
        self.exit_control.start()
```

```python
fake_csc = FakeSchedulerCSC()
```
