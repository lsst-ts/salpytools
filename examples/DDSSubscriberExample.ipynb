```python
from lsst.ts.salpytools import salpylib
```

```python
#1) Create a dictionary for our subscribers to store data to.
bulk_cloud_handle = {}            # <-- a telemetry topic
invalidate_target_handle = {}     # <-- a event topic
```

```python
#2) Create the DDSSubscriber objects passing in the handle.
bulk_cloud_subscriber = salpylib.DDSSubscriberThread(topic = "scheduler_bulkCloud", handle = bulk_cloud_handle)
invalidate_target_subscriber = salpylib.DDSSubscriberThread(topic = "scheduler_logevent_invalidateTarget", 
                                                            handle = invalidate_target_handle)
```

```python
#3) Configure the DDSSubscriber object.
bulk_cloud_subscriber.configure()
invalidate_target_subscriber.configure()
```

```python
#3) Start the DDSSubscriber thead.
bulk_cloud_subscriber.start()
invalidate_target_subscriber.start()
```

```python
"""4) Now open the jupyter notebook 'TelemetryPublishers', send over a bulkCloud Topic.
and an invalidateTarget topic. Then execute this block, you should see the data here.
"""
print(bulk_cloud_handle)
print(invalidate_target_handle)
```
