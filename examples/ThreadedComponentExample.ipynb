```python
from salpytools import salpylib
```

```python
"""1) Say that this jupyter notebook is our application. We wish to reciece
data from SAL using salpylib so we can save ourselves all the threading work. 
We need a handle to pass to the library wich it will use to attach data to,
so that we have access to it here.
"""
class BulkCloudHandle():
    
    def __init__(self):
        self.bulkCloud = None
        self.timestamp = None
        
bulkCloudHandle = BulkCloudHandle()
```

```python
"""2) We create a subscriberThread object, and pass in the handle we made, along
with the topic that we want it to pull data from
"""
bulkCloudSubscriber = salpylib.subscriberThread("scheduler_bulkCloud", bulkCloudHandle)
```

```python
"3) We tell the bulkCloudSubsciber to fire begin listening for new data over SAL"
bulkCloudSubscriber.start()
```

```python
"""4) Now open the jupyter notebook 'TelemetryPublishers', send over a bulkCloud Topic.
The bulkdCloudSubsciber should have stores this newly recieved data to our handle object.
Let's print out our object to see if it saved to it correctly. Change the values, and run
this block again. You should see these changes carry through.
"""
print("bulkdCloud value: {}".format(bulkCloudHandle.bulkCloud))
print("timestamp value: {}".format(bulkCloudHandle.timestamp))
```
